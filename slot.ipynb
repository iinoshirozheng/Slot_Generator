```python
# 變數設定

import pandas
import numpy as np

csv_files = [
            r'test/record_2023_03_06_15_56_39_363.csv'
            ]

csv_datas  = []
mode_datas = []
# 讀取所有csv檔案
for file in csv_files:
    print(file)
    csv_datas.append(pandas.read_csv(file))

# 新增一個字典將 MODE 分開
for df in csv_datas:
    mode_dict = {}
    for mode in df['Mode']:
        mode_df = df[df['Mode'] == mode].copy()
        mode_df = mode_df.drop('Mode', axis = 1)
        mode_dict[mode] = mode_df
    mode_datas.append(mode_dict)
```

```python
# 顯示 第一組 Normal Data 轉成只有Label
for m_dict in mode_datas:
    m_dict["NORMAL"] = m_dict["NORMAL"].filter(regex='^L')
    m_dict["FREE"] = m_dict["FREE"].filter(regex='^L')

# 查看第一組 Normal Data 轉成只有 Label 的資料
mode_datas[0]["NORMAL"]
```

```python
# 定義一個轉換字典將所有資料轉成編號，用於之後方便計算
label_map = {"A":0, "J":1, "Q":2, "K":3, "Boy":4, "Girl":5, "WILD":6, "Wild":6, "Scatter":7, "Carpet":8}
label_list = ["A","J", "Q", "K", "Boy", "Girl" , "Wild", "Scatter", "Carpet"]

for mode_data in mode_datas:
    mode_data["NORMAL"] = mode_data["NORMAL"].applymap(lambda x: label_map.get(x, x))
    mode_data["FREE"] = mode_data["FREE"].applymap(lambda x: label_map.get(x, x))

# 顯示第一組 的 轉換 資料 (NORMAL)
mode_datas[0]["NORMAL"]
```

```python
# 轉成矩陣

for mode_data in mode_datas:
    mode_data["NORMAL"] = mode_data["NORMAL"].to_numpy().reshape(-1, 5, 3)
    mode_data["FREE"] = mode_data["FREE"].to_numpy().reshape(-1, 5, 3)

```

```python
def count_labels(L, name):
    label_arr, counts = np.unique(L, axis=0, return_counts=True)

    # Find the indices of label categories with more than one item
    indices = np.where(counts > 1)[0]
    
    # Sort the label categories by count (from high to low) and keep only those with more than one item
    sorted_indices = np.argsort(-counts[indices])
    sorted_counts = counts[indices][sorted_indices]
    sorted_labels = [label_arr[i] for i in indices[sorted_indices]]
    
    num = 0
    for label, count in zip(sorted_labels, sorted_counts):
        print(f"{label} appears {count} items.\t{count/sum(counts)*100:.2f}% ({count}/{sum(sorted_counts)})")
        num += 1
        
    print(f"L{name}: {num} categories with more than one item.")
    print()
    
    return sorted_counts/sum(sorted_counts)*100


count_labels(mode_datas[0]["NORMAL"][:, 0], 1)
#將資料存為百分比

mode_percentages = []

for data in mode_datas:
    mode_percentages.append([{"NORMAL":count_labels(data["NORMAL"][:, L], L+1), "FREE":count_labels(data["FREE"][:, L], L+1)} for L in range(5)])
    
```

```python
# 設定基準值 MG FG
MG_Percent = [60, 61, 61, 53, 53]
FG_Percent = [50, 50, 50, 62, 50]
```

```python
# 計算均方差

def MSE(data, baseline):
    return np.square(data - baseline).mean(axis=0)

normal_mse_percentages = []
free_mse_percentages = []

for mode in mode_percentages:
    for index, mg in enumerate(MG_Percent):
        normal_mse_percentages.append( {"NORMAL" : MSE(mode[index]["NORMAL"], (1/float(mg))*100.0)} )

for mode in mode_percentages:
    for index, mg in enumerate(FG_Percent):
        free_mse_percentages.append( {"FREE": MSE(mode[index]["FREE"], (1/float(mg))*100.0)} )

normal_mse_percentages
```

```python
def transform_matrix(labels_list):
    transform_arr = np.zeros((len(label_list), len(label_list)), dtype=float)
    for data,count in labels_list:
        transform_arr[data[0]][data[1]] += float(count)
        transform_arr[data[1]][data[2]] += float(count)
    # Normalize the rows of the array
    row_sums = transform_arr.sum(axis=1)
    nonzero_row_indices = row_sums.nonzero()[0]
    transform_arr[nonzero_row_indices] /= row_sums[nonzero_row_indices][:, np.newaxis]
    return transform_arr

def generate_label(list_len, limit = False):
    data = zip(sorted_labels, sorted_counts)
    matrix = transform_matrix(data)
    # 選一個連續 label 並轉成列表
    random = np.random.choice(len(sorted_labels))
    gen_list = []
    for label in sorted_labels[random]:
        gen_list.append(label)
    check_list = [True for i in range(len(sorted_labels))] if limit else [False for i in range(len(sorted_labels))]
    check_list[random] = False


    # 當小於指定長度就一直生成
    while sum(check_list) or (len(gen_list) <=  list_len):

        # 抓取最後一個並且映射到轉置矩陣之中
        current_arr = matrix[gen_list[-1]]
        # 或取下一個要接的 LIST 的索引值
        next_index = np.random.choice(len(current_arr), p=current_arr)

        if limit :
            for idx, label in enumerate(sorted_labels):
                if np.array_equal(label, [gen_list[-2], gen_list[-1], int(next_index)]):
                    for label in sorted_labels:
                        if np.array_equal(label[:2], [gen_list[-1], int(next_index)]):
                            gen_list.append(int(next_index))
                            check_list[idx] = False
                            # if len(gen_list) >= list_len : 
                            #     return []
                            break
                    break
        else:
            gen_list.append(int(next_index))

    return gen_list

print(len(generate_label(0, True)))
```
